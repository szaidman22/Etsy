## Maine

```python
#Maine: https://www.etsy.com/search/vintage/clothing/womens-clothing?q=womens+vintage+clothing&locationQuery=4971068&explicit=1&ref=pagination&page=2
        
listing_info = []
for i in range(1,84):      # Number of pages plus one 
    url = "https://www.etsy.com/search/vintage/clothing/womens-clothing?q=womens+vintage+clothing&locationQuery=4971068&explicit=1&ref=pagination&page={}".format(i)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    results = soup.find(id="content")
    job_elems = results.find_all(class_= 'organic-impression display-inline-block listing-link ')
    links_stage = []
   # link_elem = []
    for job_elem in job_elems:
        title_elem = job_elem.find(class_='text-gray text-truncate mb-xs-0 text-body')
        title_elem = bleach.clean(title_elem, tags=[], strip=True).strip()
        price_elem = job_elem.find(class_='currency-value')
        price_elem = bleach.clean(price_elem, tags=[], strip=True).strip()
       # links_stage.append(job_elem)
       # for x in links_stage:
          #  link_elem = (x.get('href'))
        listing_info.append((title_elem,price_elem,"Maine", "Maine","Northeast"))
    print(len(listing_info))

columns = ['Description','Price','City','State','Region']
listing_data = pd.DataFrame(listing_info,columns=columns)
maine_listings = listing_data
maine_listings.head()
    
```
