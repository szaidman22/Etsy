```python
import requests
from bs4 import BeautifulSoup
import bleach
import pprint
import pandas as pd
```

```python
#try looping over multiple web pages
#THIS WORKS!!!! THIS IS GOOD!!!
#add piece that automatically stops when all listings are done

listing_info = []
for i in range(1,2):      # Number of pages plus one 
    url = "https://www.etsy.com/search/vintage/clothing/womens-clothing?q=womens+vintage+clothing&locationQuery=5332921&explicit=1&ref=pagination&page={}".format(i)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    results = soup.find(id="content")
    job_elems = results.find_all(class_= 'organic-impression display-inline-block listing-link ')
    links_stage = []
    link_elem = []
    for job_elem in job_elems:
        title_elem = job_elem.find(class_='text-gray text-truncate mb-xs-0 text-body')
        title_elem = bleach.clean(title_elem, tags=[], strip=True).strip()
        price_elem = job_elem.find(class_='currency-value')
        price_elem = bleach.clean(price_elem, tags=[], strip=True).strip()
        links_stage.append(job_elem)
        for x in links_stage:
            link_elem = (x.get('href'))
            page = requests.get((x.get('href')))
            soup = BeautifulSoup(page.content, 'html.parser')
            results = soup.find(id="content")  
            place_stage = results.find(class_="display-inline-block text-body text-gray-lightest")
            place_elem = bleach.clean(place_stage, tags=[], strip=True).strip()
            
        listing_info.append((title_elem,price_elem,"Berkeley, CA", link_elem, place_elem))
    print(len(listing_info))
```
